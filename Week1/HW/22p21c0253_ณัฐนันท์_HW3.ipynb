# Histogram

## Age

```
df_age = df['age']
plt.style.use('ggplot')
plt.hist(df_age, bins=20)
plt.show()
```

## Creatinine phosphokinase

Level of the CPK enzyme in the blood (mcg/L)

```
df_creatinine_phosphokinase = df['creatinine_phosphokinase']
plt.style.use('ggplot')
plt.hist(df_age, bins=20)
plt.show()
```

##Platelets
Platelets in the blood (kiloplatelets/mL)

```
df_platelets = df['platelets']
plt.style.use('ggplot')
plt.hist(df_age, bins=20)
plt.show()
```

##Serum Creatinine

```
df_serum_creatinine = df['serum_creatinine']
plt.style.use('ggplot')
plt.hist(df_age, bins=20)
plt.show()
```

##Serum sodium
Level of serum sodium in the blood (mEq/L)

```
df_serum_sodium = df['serum_sodium']
plt.style.use('ggplot')
plt.hist(df_age, bins=20)
plt.show()
```

# Bar plot

##Anaemia
Decrease of red blood cells or hemoglobin

```
df_anaemia = df['anaemia'].value_counts().to_frame()
df_anaemia['name'] = df_anaemia.index
df_anaemia = df_anaemia.reset_index()
df_anaemia = df_anaemia.drop(columns=['index'])
```

```
df_anaemia.plot(kind="bar", x="name", y="anaemia")
# 1 เป็น
# 0 ไม่เป็น
```

##Diabetes
If the patient has diabetes


```
df_diabetes = df['diabetes'].value_counts().to_frame()
df_diabetes['name'] = df_diabetes.index
df_diabetes = df_diabetes.reset_index()
df_diabetes = df_diabetes.drop(columns=['index'])
```

```
df_diabetes.plot(kind="bar", x="name", y="diabetes")
# 1 เป็น
# 0 ไม่เป็น
```

##High Blood Pressure
If the patient has hypertension

```
df_high_blood_pressure = df['high_blood_pressure'].value_counts().to_frame()
df_high_blood_pressure['name'] = df_high_blood_pressure.index
df_high_blood_pressure = df_high_blood_pressure.reset_index()
df_high_blood_pressure = df_high_blood_pressure.drop(columns=['index'])
```

```
df_high_blood_pressure.plot(kind="bar", x="name", y="high_blood_pressure")
# 1 เป็น
# 0 ไม่เป็น
```

##Sex

```
df_sex = df['sex'].value_counts().to_frame()
df_sex['name'] = df_sex.index
df_sex = df_sex.reset_index()
df_sex = df_sex.drop(columns=['index'])
```

```
df_sex.plot(kind="bar", x="name", y="sex")
# 1 เป็น
# 0 ไม่เป็น
```

##Smoking

```
df_smoking = df['smoking'].value_counts().to_frame()
df_smoking['name'] = df_smoking.index
df_smoking = df_smoking.reset_index()
df_smoking = df_smoking.drop(columns=['index'])
```
